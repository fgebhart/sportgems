```python
import folium
from folium import plugins
from pathlib import Path
import numpy as np
from sportgems import find_fastest_section_in_fit, parse_fit_data, find_fastest_section
```

```python
fit_file = Path(".").parent / "data" / "2019-09-14-17-22-05.fit"
```

```python
fit_data = parse_fit_data(str(fit_file))
```

```python
coords = []
for coordinate in fit_data.coordinates:
    if coordinate[0] > 0 and coordinate[1] > 0:
        coords.append((coordinate[0], coordinate[1]))
```

```python
fs_one_km = find_fastest_section(1000, fit_data.times, fit_data.coordinates)
one_km_coords = coords[fs_one_km.start_index:fs_one_km.end_index]
```

```python
fs_two_km = find_fastest_section(2000, fit_data.times, fit_data.coordinates)
two_km_coords = coords[fs_two_km.start_index:fs_two_km.end_index]
```

```python
fs_three_km = find_fastest_section(3000, fit_data.times, fit_data.coordinates)
three_km_coords = coords[fs_three_km.start_index:fs_three_km.end_index]
```

```python
fs_five_km = find_fastest_section(5000, fit_data.times, fit_data.coordinates)
five_km_coords = coords[fs_five_km.start_index:fs_five_km.end_index]
```

```python
m = folium.Map(location=fit_data.coordinates[300], zoom_start=15)

trace = folium.PolyLine(coords, color="red")
fs_one_km = folium.PolyLine(one_km_coords, color="yellow")
fs_two_km = folium.PolyLine(two_km_coords, color="blue")
fs_three_km = folium.PolyLine(three_km_coords, color="green")
fs_five_km = folium.PolyLine(five_km_coords, color="black")

trace.add_to(m)
fs_five_km.add_to(m)
fs_three_km.add_to(m)
fs_two_km.add_to(m)
fs_one_km.add_to(m)
m
```
